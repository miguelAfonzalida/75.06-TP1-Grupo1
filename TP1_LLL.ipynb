# Detectando hashtags y menciones

```python
regex = '\s([@#][\w_-]+)'
```

```python
textos = df['text']
resultados = []
for t in textos:
    matched = re.findall(regex,t)
    if (len(matched) > 0):
        resultados.append(matched)
resultados
```

```python
hashtags = []
mentions = []
for i in resultados:
    for j in i:
        if ('@' in j):
            mentions.append(j)
        else:
            hashtags.append(j)
```

```python
mentions_for_wordcloud = " ".join(str(each) for each in mentions)
hashtags_for_wordcloud = " ".join(str(each) for each in hashtags)
```

# Limpieza de tweets

```python
tok = WordPunctTokenizer()
#regular expression for eliminate @person
pat1 = r'\s([@#][\w_-]+)'
#regular expression for eliminate links
pat2 = r'https?://[A-Za-z0-9./]+'
combined_pat = r'|'.join((pat1, pat2))

def tweet_cleaner(text):
    #dealing with html encoding
    soup = BeautifulSoup(text, 'lxml')
    souped = soup.get_text()
    #applying pattern regexs
    stripped = re.sub(pat2, '', souped)
    try:
        clean = stripped.decode("utf-8-sig").replace(u"\ufffd", "?")
    except:
        clean = stripped
    letters_only = re.sub("[^a-zA-Z]", " ", clean)
    lower_case = letters_only.lower()
    # During the letters_only process two lines above, it has created unnecessay white spaces,
    # I will tokenize and join together to remove unneccessary white spaces
    words = tok.tokenize(lower_case)
    return (" ".join(words)).strip()
```
